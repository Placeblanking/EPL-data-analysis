```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
epl_df = pd.read_csv('/content/EPL_20_21.csv')
epl_df.head()
```

```python
epl_df.info()
```

```python
epl_df.describe()
```

```python
epl_df.isna().sum()
```

```python
epl_df['MinsPerMatch'] = (epl_df['Mins']/epl_df['Matches']).astype(int)
epl_df['GoalsPerMatch'] = (epl_df['Goals']/epl_df['Matches']).astype(float)
print(epl_df.head())
```

```python
Total_Goals = epl_df['Goals'].sum()
print(Total_Goals)
```

```python
Total_Penalty_Goals = epl_df['Penalty_Goals'].sum()
print(Total_Penalty_Goals)
```

```python
Total_PenaltyAttempts = epl_df['Penalty_Attempted'].sum()
print(Total_PenaltyAttempts)
```

```python
plt.figure(figsize=(13, 6))
pl_not_scored = epl_df['Penalty_Attempted'].sum() - Total_Penalty_Goals
data = [pl_not_scored, Total_Penalty_Goals]
labels = ['Penalties misseed', 'Penalties Scored']
color = sns.color_palette('Set2')
plt.pie(data, labels=labels, colors=color, autopct='%.0f%%')
plt.show()
```

```python
epl_df['Club'].value_counts().nsmallest(5).plot(kind='bar', color=sns.color_palette("viridis"))
```

```python
Under20 = epl_df[epl_df['Age'] <= 20]
age20_25 = epl_df[(epl_df['Age'] > 20) & (epl_df['Age'] <= 25)]
age25_30 = epl_df[(epl_df['Age'] > 25) & (epl_df['Age'] <= 30)]
age30 = epl_df[epl_df['Age'] > 30]
```

```python
x = np.array([Under20['Name'].count(), age20_25['Name'].count(), age25_30['Name'].count(), age30['Name'].count()])
mylabels = ["Under 20", "20-25", "25-30", "Above 30"]
plt.pie(x, labels = mylabels, autopct="%.1f%%")
plt.show()
```

```python
players_under_20 = epl_df[epl_df['Age'] < 20]
players_under_20['Club'].value_counts().plot(kind = 'bar', color = sns.color_palette("cubehelix"))
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Club', y='Age', data=epl_df)
plt.xticks(rotation=90)
plt.show()
```

```python
num_player = epl_df.groupby('Club').size() # Call the size method to get the actual size
data = (epl_df.groupby('Club')['Age'].sum())/ num_player
data.sort_values(ascending = False)
```

```python
Assists_by_club = pd.DataFrame(epl_df.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style="whitegrid",color_codes=True)
ax = sns.barplot(x="Club", y="Assists", data=Assists_by_club.sort_values(by='Assists'),palette='Set2') # Changed Assists_by_clubs to Assists_by_club
ax.set_xlabel('Club',fontsize=20) # Changed FontSize to fontsize
ax.set_ylabel('Assists',fontsize=20) # Changed FontSize to fontsize
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"] = (20,8)
plt.title('Plot of Clubs vs Total Assists',fontsize=20)
```

```python
top_10_assists = epl_df.sort_values(by='Assists', ascending=False).head(10)
top_10_assists = top_10_assists[['Name', 'Club', 'Assists', 'Matches']]
top_10_assists
```

```python
club_goals = epl_df.groupby('Club')['Goals'].sum()

plt.figure(figsize=(12, 6))
club_goals.plot(kind='bar', color=sns.color_palette("viridis"))
plt.title('Total Goals per Club')
plt.xlabel('Club')
plt.ylabel('Total Goals')
plt.xticks(rotation=75)
plt.show()
```

```python
top_10_goals = epl_df.sort_values(by='Goals', ascending=False).head(10)
top_10_goals = top_10_goals[['Name', 'Club', 'Goals', 'Matches']]
top_10_goals
```

```python
top_10_goals = epl_df.sort_values(by='Goals', ascending=False).head(10)
top_10_goals['GoalsPerMatch'] = (top_10_goals['Goals'] / top_10_goals['Matches']).astype(float)
top_10_goals = top_10_goals[['Name', 'GoalsPerMatch', 'Matches', 'Goals']]
top_10_goals
```

```python
plt.figure(figsize=(14, 7))
assists = epl_df['Assists'].sum()
data = [Total_Goals - assists, assists]
labels = ['Goals without assists', 'Goals with assists']
color = sns.color_palette('Set1')
plt.pie(data, labels=labels, colors=color, autopct='%.0f%%')
plt.show()
```
